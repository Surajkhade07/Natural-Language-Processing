```python
import pandas as pd
```

```python
# load data
message=pd.read_csv('SMSSpamCollection.csv')
```

```python
message
```

```python
message.shape
```

```python
#data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range (0,len(message)):
    clean_data=re.sub('[^a-zA-Z0-9]',' ',message['message'][i])
    claen_data=clean_data.lower()
    clean_data=claen_data.split()
    clean_data=[ps.stem(word) for word in clean_data if not word in (stopwords.words('english'))]
    clean_data=' '.join(clean_data)
    corpus.append(clean_data)
```

```python
corpus[1:5]
```

```python
# use of bag of word 
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,binary=True,ngram_range=(2, 2))
X=cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
Y=pd.get_dummies(message['label'])
Y=Y.iloc[:,1].astype(int).values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=30)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detection_model=MultinomialNB()
spam_detection_model.fit(X_train,Y_train)
```

```python
#prediction 
Y_pred=spam_detection_model.predict(X_test)
```

```python
Y_pred
```

```python
#check accuracy
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(Y_test,Y_pred)
print("Acuuracy:",score)
report=classification_report(Y_test,Y_pred)
print("report:",report)
```

```python
# see diffrance using confusion metrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_pred)
```

```python
cm
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(max_features=2500,binary=True)
X1=tf.fit_transform(corpus).toarray()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X1,Y,test_size=0.2,random_state=30)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detection_model=MultinomialNB()
spam_detection_model.fit(X_train,Y_train)
```

```python
#prediction 
Y_pred=spam_detection_model.predict(X_test)
```

```python
#check accuracy
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(Y_test,Y_pred)
print("Acuuracy:",score)
report=classification_report(Y_test,Y_pred)
print("report:",report)
```
